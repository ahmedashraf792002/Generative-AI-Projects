```python
### Import Libraries
import os
import json
import nest_asyncio
from llama_index.core import SimpleDirectoryReader
from llama_index.core import SummaryIndex
from llama_index.llms.groq import Groq
from llama_index.core.node_parser import SentenceSplitter
from llama_index.core import Settings
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
nest_asyncio.apply()
import warnings
warnings.filterwarnings('ignore')
```

```python
### GROQ API Config
config_data = json.load(open("config.json"))
GROQ_API_KEY = config_data["api_key"]
os.environ["GROQ_API_KEY"] = GROQ_API_KEY
```

```python
### Load Llama3 Model & HuggingFaceEmbedding
llm = Groq(model="llama-3.1-8b-instant")
Settings.llm = llm
Settings.embed_model = HuggingFaceEmbedding()
```

```python
# load the document
documents = SimpleDirectoryReader(input_files=["Lecture_7-unsupervised-learning.pdf"]).load_data()
```

```python
print(type(documents),len(documents))
```

```python
# Splitting large texts into smaller chunks
### Initialize SentenceSplitter with a chunk size of 2048 characters ###
splitter = SentenceSplitter(chunk_size=2048)
# Extracting nodes from the documents
### Convert documents into nodes for further processing ###
nodes = splitter.get_nodes_from_documents(documents)
```

```python
# Create an index for summarization
summary_index = SummaryIndex(nodes)
```

```python
# Convert it into a query engine for summarization
summary_query_engine = summary_index.as_query_engine(
    response_mode="tree_summarize",
    use_async=True
)
```

```python
# Query the engine to generate a summary
def Query(query):
    response = summary_query_engine.query(query)
    print("Query : ", query)
    print("\n","*"*80,"\n")
    print("Response : ",response)
```

```python
Query("Summarize the main points of the document")
```

```python
Query("Give a summary of the given document")
```

```python
Query("Give a detailed summary of the given document")
```

```python
Query("Give a summary of kmeans")
```

```python
Query("Give a summary of hierarchical clustering, and density-based clustering")
```

```python
# storing the document summary index
summary_index.storage_context.persist("doc_summary_index")
```
